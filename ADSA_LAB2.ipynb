1. Sum of Elements in an Array
• Scenario: You have a list of daily sales of a store.
* Write a simple iterative program to compute the sum of elements.
* Write a recursive version of the same program.


```python
%%writefile 1a.c
#include <stdio.h>

int main() {
    int arr[] = {5, 10, 15, 20, 25};
    int n = sizeof(arr)/sizeof(arr[0]);
    int sum = 0;

    for(int i = 0; i < n; i++) {
        sum += arr[i];
    }

    printf("Sum = %d\n", sum);
    return 0;
}

```

```python
!gcc 1a.c -o 1a
!./1a
```

```python
%%writefile 1b.c
#include <stdio.h>
int sum(int arr[], int n) {
    if(n == 0) return 0;
    return arr[n-1] + sum(arr, n-1);
}

int main() {
    int arr[] = {5, 10, 15, 20, 25};
    int n = sizeof(arr)/sizeof(arr[0]);
    printf("Sum = %d\n", sum(arr, n));
    return 0;
}
```

2. Find Maximum Element in an Array
• Scenario: Given temperature readings for a week.
* Find maximum element using a simple linear scan (O(n)).
* Can you find it by first sorting the array and taking the last element?

```python
%%writefile 2a.c
#include <stdio.h>
int arr[] = {5, 10, 15, 20, 25,5,6,7,8,9,12,56,19,90,33,40,50};
int n = sizeof(arr)/sizeof(arr[0]);
int main(){
    int max = arr[0];
    for(int i = 1; i < n; i++) {
        if(arr[i] > max) {
            max = arr[i];
        }
    }
    printf("Max = %d\n", max);
    return 0;
}
```

```python
%%writefile 2b.c
#include <stdio.h>

int main() {
    int arr[] = {32, 35, 31, 37, 30, 36, 34};
    int n = sizeof(arr)/sizeof(arr[0]);

    for(int i = 0; i < n-1; i++) {
        for(int j = 0; j < n-i-1; j++) {
            if(arr[j] > arr[j+1]) {
                int temp = arr[j];
                arr[j] = arr[j+1];
                arr[j+1] = temp;
            }
        }
    }

    printf("Maximum = %d\n", arr[n-1]);
    return 0;
}
```

3. Reverse a String
• Scenario: Reversing a username for some encryption purpose.
* Reverse in-place using two-pointer technique (O(n), O(1)).
* Reverse by creating a new array (O(n), O(n)).

```python
%%writefile 3a.c
#include<stdio.h>
int main(){
    char str[] = "Hello World";
    int n = sizeof(str)/sizeof(str[0]);
    int start = 0;
    int end = n-2;
    while(start < end){
        char temp = str[start];
        str[start] = str[end];
        str[end] = temp;
        start++;
        end--;
    }
    printf("%s\n", str);
}
```

5. Factorial of a Number.
Scenario: Calculate the number of possible arrangements.
* Iterative method (O(n), O(1)).
* Recursive method (O(n), O(n) for call stack).


```python
%%writefile 5a.c
#include <stdio.h>
int main(){
  int n ;
  scanf("%d", &n);
  int fact = 1;
  for(int i = 1; i <= n; i++) {
    fact *= i;
  }
  printf("Factorial of %d is %d\n", n, fact);
}
```

```python
%%writefile 5b.c
#include <stdio.h>
int fact(int n) {
  if(n == 0) return 1;
  return n * fact(n-1);
}
int main(){
  int n;
  scanf("%d",&n);
  printf("Factorial of %d is %d\n", n, fact(n));
}


```

6. Linear Search
• Scenario: Search for a customer ID in a small dataset.
* Implement basic linear search (O(n), O(1)).
* Optimize by using sentinel method to reduce comparisons (still O(n), but fewer  operations).


```python
%%writefile 6a.c
#include<stdio.h>
int main(){
  int arr[] = {32, 35, 31, 37, 30, 36, 34};
  int n = sizeof(arr)/sizeof(arr[0]);
  int id;
  scanf("%d", &id);
  for(int i = 0; i < n; i++) {
    if(arr[i] == id) {
      printf("Found at index %d\n", i);
      return 0;
    }
  }
  printf("Not found\n");

}
```

```python
!gcc 6a.c -o 6a
!./6a
```

```python
%%writefile 6b.c
#include<stdio.h>
int sentinelSearch(int arr[], int n, int key){
  int last = arr[n-1];
  arr[n-1] = key;

  int i = 0;
  while(arr[i] != key)
  i++;

  arr[n-1] = last;
  if(i < n-1 || arr[n-1] == key)
  return i;
  else
  return -1;


}

int main(){
  int arr[] = {32, 35, 31, 37, 30, 36, 34};
  int n = sizeof(arr)/sizeof(arr[0]);
  int id;
  scanf("%d", &id);
  int index = sentinelSearch(arr, n, id);
  if(index == -1)
  printf("Not found\n");
  else
  printf("Found at index %d\n", index);
  return 0;
}
```

* Traverse string and check each character (O(n), O(1)).
* Use a lookup table (array of 256 size) to speed up vowel checking (O(n), O(1) but extra  space).


```python
%%writefile 8a.c
#include<stdio.h>
int main(){
  char c[] = "Analyze user comments to count number of vowels.";
  int n = sizeof(c)/sizeof(c[0]);
  int count = 0;
  for(int i = 0; i < n; i++) {
    if(c[i]=='a'||c[i]=='e'||c[i]=='i'||c[i]=='o'||c[i]=='u'||c[i]=='A'||c[i]=='E'||c[i]=='I'||c[i]=='O'||c[i]=='U')
    count++;
  }
  printf("Number of vowels in the string are %d\n", count);
}
```

10. Check Palindrome Number
• Scenario: Validate identification number (ID) is symmetric.
* Convert number to string and check (O(n), O(n)).
* Mathematical method: Reverse digits without converting (O (log n), O (1)).


```python
%%writefile 10a.c
#include<stdio.h>
#include<string.h>


int isPalindrome(int num) {
  char c[20];
  sprintf(c, "%d", num);
  int len = strlen(c);
  int count = 0;
  for(int i = 0; i < len; i++) {
    if(c[i]==c[len-i-1])
    count++;
  }
  return count == len;
}

int main(){
  int n = 123321;
  int m = 4356;

  if (isPalindrome(n)) {
    printf("%d is a palindrome\n", n);
  } else {
    printf("%d is not a palindrome\n", n);
  }

  if (isPalindrome(m)) {
    printf("%d is a palindrome\n", m);
  } else {
    printf("%d is not a palindrome\n", m);
  }

  return 0;
}
```

```python
!gcc 10a.c -o 10a
!./10a
```

```python
%%writefile 10b.c
#include<stdio.h>
int main(){
  int a = 123321;

  int rev = 0;
  int temp = a;
  while(temp > 0) {
    rev = rev * 10 + temp % 10;
    temp /= 10;
  }
  if(a == rev)
  printf("%d is a palindrome\n", a);
  else
  printf("%d is not a palindrome\n", a);
}
```
