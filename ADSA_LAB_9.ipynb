1. Write a program to construct a binary tree using user input (level-order insertion).
Perform preorder, inorder, and postorder traversals.

```python
%%writefile 1.c
#include <stdio.h>
#include <stdlib.h>

struct Node {
    int data;
    struct Node* left;
    struct Node* right;
};

// Simple queue for Node*
struct Node* queueArr[200];
int frontIdx = -1, rearIdx = -1;

void enqueue(struct Node* x) {
    if (rearIdx == 199) return;
    if (frontIdx == -1) frontIdx = 0;
    queueArr[++rearIdx] = x;
}

struct Node* dequeue() {
    if (frontIdx == -1 || frontIdx > rearIdx) return NULL;
    return queueArr[frontIdx++];
}

void resetQueue() { frontIdx = rearIdx = -1; }

struct Node* newNode(int value) {
    struct Node* temp = (struct Node*)malloc(sizeof(struct Node));
    temp->data = value;
    temp->left = temp->right = NULL;
    return temp;
}

// Level-order insertion
struct Node* insert(struct Node* root, int value) {
    struct Node* temp = newNode(value);
    if (root == NULL) return temp;

    // use queue to find first empty spot
    resetQueue();
    enqueue(root);

    while (frontIdx <= rearIdx) {
        struct Node* curr = dequeue();
        if (curr->left == NULL) {
            curr->left = temp;
            return root;
        } else enqueue(curr->left);

        if (curr->right == NULL) {
            curr->right = temp;
            return root;
        } else enqueue(curr->right);
    }
    return root;
}

// Level-order traversal (for verification)
void printLevelOrder(struct Node* root) {
    if (root == NULL) return;
    resetQueue();
    enqueue(root);
    while (frontIdx <= rearIdx) {
        struct Node* curr = dequeue();
        printf("%d ", curr->data);
        if (curr->left) enqueue(curr->left);
        if (curr->right) enqueue(curr->right);
    }
}

// ---- Recursive traversals ----
void inorder(struct Node* root) {
    if (root == NULL) return;
    inorder(root->left);
    printf("%d ", root->data);
    inorder(root->right);
}

void preorder(struct Node* root) {
    if (root == NULL) return;
    printf("%d ", root->data);
    preorder(root->left);
    preorder(root->right);
}

void postorder(struct Node* root) {
    if (root == NULL) return;
    postorder(root->left);
    postorder(root->right);
    printf("%d ", root->data);
}

int main() {
    struct Node* root = NULL;

    // sample inserts
    root = insert(root, 10);
    root = insert(root, 20);
    root = insert(root, 30);
    root = insert(root, 40);
    root = insert(root, 50);

    printf("Level Order: ");
    printLevelOrder(root);
    printf("\n");

    printf("Inorder: ");
    inorder(root);
    printf("\n");

    printf("Preorder: ");
    preorder(root);
    printf("\n");

    printf("Postorder: ");
    postorder(root);
    printf("\n");

    return 0;
}
```

2. Given a binary tree, write functions to compute:
 -  Height of the tree
 - Total number of leaf nodes

```python
%%writefile 2.c
#include <stdio.h>
#include <stdlib.h>

// ---------------- NODE STRUCT ----------------
struct Node {
    int data;
    struct Node* left;
    struct Node* right;
};

// Create new node
struct Node* newNode(int value) {
    struct Node* temp = (struct Node*)malloc(sizeof(struct Node));
    temp->data = value;
    temp->left = temp->right = NULL;
    return temp;
}

// ------------- HEIGHT OF TREE ----------------
int height(struct Node* root) {
    if (root == NULL)
        return 0;

    int lh = height(root->left);
    int rh = height(root->right);

    return 1 + (lh > rh ? lh : rh);
}

// ------------- COUNT LEAF NODES --------------
int countLeaves(struct Node* root) {
    if (root == NULL)
        return 0;

    if (root->left == NULL && root->right == NULL)
        return 1;

    return countLeaves(root->left) + countLeaves(root->right);
}

// -------------------- MAIN --------------------
int main() {

    // Creating a simple binary tree manually:
    struct Node* root = newNode(10);
    root->left = newNode(20);
    root->right = newNode(30);
    root->left->left = newNode(40);

    printf("Height of Tree: %d\n", height(root)-1);
    printf("Total Leaf Nodes: %d\n", countLeaves(root));

    return 0;
}
```

3. Given any binary tree, write a program to verify whether it satisfies BST properties using the
minâ€“max recursive method.

```python
%%writefile 3_verify_bst.c
#include <stdio.h>
#include <stdlib.h>
#include <limits.h>

typedef struct Node {
    int data;
    struct Node *left, *right;
} Node;

Node* newNode(int val){
    Node* n = (Node*)malloc(sizeof(Node));
    n->data = val; n->left = n->right = NULL; return n;
}

int isBSTUtil(Node* root, int minv, int maxv){
    if(!root) return 1;
    if(root->data <= minv || root->data >= maxv) return 0;
    return isBSTUtil(root->left, minv, root->data) && isBSTUtil(root->right, root->data, maxv);
}

int isBST(Node* root){
    return isBSTUtil(root, INT_MIN, INT_MAX);
}

/* helper to build a tree from input level-wise (-1 = NULL) */
Node* buildFromLevel(){
    int x;
    if(scanf("%d", &x)!=1) return NULL;
    if(x==-1) return NULL;
    Node** q = malloc(sizeof(Node*) * 10000);
    int head=0, tail=0;
    Node* root = newNode(x);
    q[tail++] = root;
    while(head<tail){
        Node* cur = q[head++];
        int l, r;
        if(scanf("%d", &l)!=1) break;
        if(l != -1){ cur->left = newNode(l); q[tail++] = cur->left;}
        if(scanf("%d", &r)!=1) break;
        if(r != -1){ cur->right = newNode(r); q[tail++] = cur->right;}
    }
    free(q);
    return root;
}

int main(){
    printf("Enter level-order node values, use -1 for NULL. First value is root.\n");
    printf("Example for tree with root 10, left 5, right 15: 10 5 15 -1 -1 -1 -1\n");
    Node* root = buildFromLevel();
    if(isBST(root)) printf("The tree IS a BST\n");
    else printf("The tree is NOT a BST\n");
    return 0;
}
```

4. Perform level-order (breadth-first) traversal of a binary tree using a queue.

```python
%%writefile 4_levelorder_fixed.c
#include <stdio.h>
#include <stdlib.h>

typedef struct Node {
    int data;
    struct Node *left, *right;
} Node;

Node* newNode(int v){
    Node* n = (Node*)malloc(sizeof(Node));
    n->data = v; n->left = n->right = NULL; return n;
}

/*
Input format:
First line: m = number of level-order slots (including -1 placeholders).
Second line: m integers giving the level-order array; use -1 to mean NULL at that position.

Example:
7
1 2 3 -1 -1 4 5

This describes the tree:
        1
       / \
      2   3
         / \
        4   5
*/
Node* buildFromLevelArray(int arr[], int m){
    if(m<=0) return NULL;
    if(arr[0]==-1) return NULL;
    Node** nodes = malloc(sizeof(Node*) * m);
    for(int i=0;i<m;i++){
        nodes[i] = (arr[i]==-1) ? NULL : newNode(arr[i]);
    }
    for(int i=0;i<m;i++){
        if(nodes[i]==NULL) continue;
        int l = 2*i + 1;
        int r = 2*i + 2;
        if(l < m) nodes[i]->left = nodes[l];
        if(r < m) nodes[i]->right = nodes[r];
    }
    Node* root = nodes[0];
    free(nodes);
    return root;
}

void levelOrder(Node* root){
    if(!root){
        printf("(empty tree)\n");
        return;
    }
    // queue implemented as dynamic array of Node*
    Node** q = malloc(sizeof(Node*) * 10000);
    int head=0, tail=0;
    q[tail++] = root;
    while(head < tail){
        Node* cur = q[head++];
        printf("%d ", cur->data);
        if(cur->left) q[tail++] = cur->left;
        if(cur->right) q[tail++] = cur->right;
    }
    printf("\n");
    free(q);
}

int main(){
    int m;
    printf("Enter number of level-order slots (m): ");
    if(scanf("%d", &m)!=1) return 0;
    if(m <= 0){ printf("m must be > 0\n"); return 0; }
    int *arr = malloc(sizeof(int)*m);
    printf("Enter %d integers (use -1 for NULL slots) in level-order:\n", m);
    for(int i=0;i<m;i++) scanf("%d", &arr[i]);

    Node* root = buildFromLevelArray(arr, m);
    printf("Level-order traversal: ");
    levelOrder(root);

    free(arr);
    return 0;
}
```

6. Given a BST, write functions to find:
* k-th smallest element
* k-th largest element

(Without storing traversal in an array)

```python
%%writefile 6_kth_both_fixed.c
#include <stdio.h>
#include <stdlib.h>
#include <stdbool.h>

// Define the structure for a BST node
typedef struct Node {
    int data;
    struct Node* left;
    struct Node* right;
} Node;

Node* createNode(int data) {
    Node* newNode = (Node*)malloc(sizeof(Node));
    if (newNode == NULL) {
        perror("Failed to allocate memory for new node");
        exit(EXIT_FAILURE);
    }
    newNode->data = data;
    newNode->left = newNode->right = NULL;
    return newNode;
}

Node* insert(Node* root, int data) {
    if (root == NULL) {
        return createNode(data);
    }
    if (data < root->data) {
        root->left = insert(root->left, data);
    } else if (data > root->data) {
        root->right = insert(root->right, data);
    }
    return root;
}

void find_kth_smallest_util(Node* root, int* k, int* result) {
    if (root == NULL || *k == 0) {
        return;
    }

    // In-order traversal (Left -> Root -> Right)
    find_kth_smallest_util(root->left, k, result);

    if (*k == 0) {
        return;
    }

    (*k)--;
    if (*k == 0) {
        *result = root->data;
        return;
    }

    find_kth_smallest_util(root->right, k, result);
}

int find_kth_smallest(Node* root, int k) {
    if (k <= 0) return -1;

    int mutable_k = k;
    int result = -1;

    find_kth_smallest_util(root, &mutable_k, &result);
    return result;
}

void find_kth_largest_util(Node* root, int* k, int* result) {
    if (root == NULL || *k == 0) {
        return;
    }

    // Reverse In-order traversal (Right -> Root -> Left)
    find_kth_largest_util(root->right, k, result);

    if (*k == 0) {
        return;
    }

    (*k)--;
    if (*k == 0) {
        *result = root->data;
        return;
    }

    find_kth_largest_util(root->left, k, result);
}

int find_kth_largest(Node* root, int k) {
    if (k <= 0) return -1;

    int mutable_k = k;
    int result = -1;

    find_kth_largest_util(root, &mutable_k, &result);
    return result;
}

void freeTree(Node* root) {
    if (root == NULL) return;
    freeTree(root->left);
    freeTree(root->right);
    free(root);
}


int main() {

    Node* root = NULL;
    root = insert(root, 50);
    insert(root, 30);
    insert(root, 70);
    insert(root, 20);
    insert(root, 40);
    insert(root, 60);
    insert(root, 80);

    // Sorted order: 20, 30, 40, 50, 60, 70, 80

    printf("BST elements in sorted order: 20, 30, 40, 50, 60, 70, 80\n");

    int k_smallest = 3;
    int result_smallest = find_kth_smallest(root, k_smallest);
    if (result_smallest != -1) {
        printf("\n%d-th smallest element is: %d (Expected: 40)\n", k_smallest, result_smallest);
    } else {
        printf("\nCould not find %d-th smallest element.\n", k_smallest);
    }

    int k_largest = 2;
    int result_largest = find_kth_largest(root, k_largest);
    if (result_largest != -1) {
        printf("%d-th largest element is: %d (Expected: 70)\n", k_largest, result_largest);
    } else {
        printf("Could not find %d-th largest element.\n", k_largest);
    }

    int k_out_of_bounds = 10;
    int result_out_of_bounds = find_kth_smallest(root, k_out_of_bounds);
    printf("\nTesting K out of bounds (K=%d):\n", k_out_of_bounds);
    if (result_out_of_bounds == -1) {
        printf("K-th smallest found: -1 (Correctly handles out of bounds)\n");
    }

    // Clean up memory
    freeTree(root);

    return 0;
}
```

7. Given a sorted array, build a balanced BST using divide-and-conquer.
Print level-order traversal of the new tree.

```python
%%writefile 7_sortedarray_to_bst.c
#include <stdio.h>
#include <stdlib.h>

typedef struct Node { int data; struct Node *left,*right; } Node;
Node* newNode(int v){ Node* n=malloc(sizeof(Node)); n->data=v; n->left=n->right=NULL; return n; }

Node* sortedToBST(int arr[], int l, int r){
    if(l>r) return NULL;
    int mid = (l + r) / 2;
    Node* root = newNode(arr[mid]);
    root->left = sortedToBST(arr, l, mid-1);
    root->right = sortedToBST(arr, mid+1, r);
    return root;
}

void levelOrder(Node* root){
    if(!root) return;
    Node** q = malloc(sizeof(Node*)*10000);
    int head=0, tail=0;
    q[tail++] = root;
    while(head<tail){
        Node* cur = q[head++];
        printf("%d ", cur->data);
        if(cur->left) q[tail++] = cur->left;
        if(cur->right) q[tail++] = cur->right;
    }
    free(q);
}

int main(){
    int n;
    printf("Enter size of sorted array: ");
    if(scanf("%d",&n)!=1) return 0;
    int *arr = malloc(sizeof(int)*n);
    printf("Enter %d sorted values (ascending):\n", n);
    for(int i=0;i<n;i++) scanf("%d",&arr[i]);
    Node* root = sortedToBST(arr, 0, n-1);
    printf("Level-order of balanced BST: ");
    levelOrder(root);
    printf("\n");
    free(arr);
    return 0;
}
```

8. Given a binary tree, write a program to check whether it is height-balanced (AVL property).
Also print the balance factor of each node.

```python
%%writefile 8_check_balanced_fixed.c
#include <stdio.h>
#include <stdlib.h>

typedef struct Node { int data; struct Node *left,*right; } Node;
Node* newNode(int v){ Node* n = malloc(sizeof(Node)); n->data=v; n->left=n->right=NULL; return n; }

/* Build tree from level-order array slots (same format as program 4) */
Node* buildFromLevelArray(int arr[], int m){
    if(m<=0) return NULL;
    if(arr[0]==-1) return NULL;
    Node** nodes = malloc(sizeof(Node*) * m);
    for(int i=0;i<m;i++){
        nodes[i] = (arr[i]==-1) ? NULL : newNode(arr[i]);
    }
    for(int i=0;i<m;i++){
        if(nodes[i]==NULL) continue;
        int l = 2*i + 1;
        int r = 2*i + 2;
        if(l < m) nodes[i]->left = nodes[l];
        if(r < m) nodes[i]->right = nodes[r];
    }
    Node* root = nodes[0];
    free(nodes);
    return root;
}

/* Postorder compute heights and print balance factor.
   Returns height of current node. 'isBalanced' set to 0 if any node violates AVL bf. */
int computeHeightAndPrintBF(Node* root, int *isBalanced){
    if(!root) return 0;
    int lh = computeHeightAndPrintBF(root->left, isBalanced);
    int rh = computeHeightAndPrintBF(root->right, isBalanced);
    int bf = lh - rh;
    printf("Node %d: left_height = %d, right_height = %d, balance_factor = %d\n", root->data, lh, rh, bf);
    if(bf > 1 || bf < -1) *isBalanced = 0;
    return 1 + (lh > rh ? lh : rh);
}

int main(){
    int m;
    printf("Enter number of level-order slots (m): ");
    if(scanf("%d",&m)!=1) return 0;
    if(m <= 0){ printf("m must be > 0\n"); return 0; }
    int *arr = malloc(sizeof(int)*m);
    printf("Enter %d integers (use -1 for NULL slots) in level-order:\n", m);
    for(int i=0;i<m;i++) scanf("%d",&arr[i]);

    Node* root = buildFromLevelArray(arr, m);
    int isBalanced = 1;
    int height = computeHeightAndPrintBF(root, &isBalanced);
    printf("Tree height = %d\n", height);
    if(isBalanced) printf("Tree IS height-balanced (AVL property holds).\n");
    else printf("Tree is NOT height-balanced (AVL property violated somewhere).\n");

    free(arr);
    return 0;
}
```

9. Implement AVL Tree insertion, after each insertion display:
* Balance Factor
* Rotation Performed

```python
%%writefile 9_avl_insertion.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>
#include <limits.h>

typedef struct Node { int data; int height; struct Node *left,*right; } Node;
Node* newNode(int v){ Node* n=malloc(sizeof(Node)); n->data=v; n->height=1; n->left=n->right=NULL; return n; }

int height(Node* n){ return n ? n->height : 0; }
int getBF(Node* n){ return n ? height(n->left) - height(n->right) : 0; }

char lastRotation[64];

Node* rightRotate(Node* y){
    Node* x = y->left;
    Node* T2 = x->right;
    x->right = y;
    y->left = T2;
    y->height = 1 + (height(y->left) > height(y->right) ? height(y->left) : height(y->right));
    x->height = 1 + (height(x->left) > height(x->right) ? height(x->left) : height(x->right));
    strcpy(lastRotation, "Right rotation (LL)");
    return x;
}

Node* leftRotate(Node* x){
    Node* y = x->right;
    Node* T2 = y->left;
    y->left = x;
    x->right = T2;
    x->height = 1 + (height(x->left) > height(x->right) ? height(x->left) : height(x->right));
    y->height = 1 + (height(y->left) > height(y->right) ? height(y->left) : height(y->right));
    strcpy(lastRotation, "Left rotation (RR)");
    return y;
}

Node* insert(Node* node, int key){
    if(!node) { strcpy(lastRotation, "No rotation"); return newNode(key); }
    if(key < node->data) node->left = insert(node->left, key);
    else if(key > node->data) node->right = insert(node->right, key);
    else { strcpy(lastRotation, "No rotation (duplicate ignored)"); return node; }

    node->height = 1 + (height(node->left) > height(node->right) ? height(node->left) : height(node->right));
    int bf = getBF(node);

    // LL
    if(bf > 1 && key < node->left->data) return rightRotate(node);
    // RR
    if(bf < -1 && key > node->right->data) return leftRotate(node);
    // LR
    if(bf > 1 && key > node->left->data){
        node->left = leftRotate(node->left);
        strcpy(lastRotation, "Left-Right rotation (LR)");
        return rightRotate(node);
    }
    // RL
    if(bf < -1 && key < node->right->data){
        node->right = rightRotate(node->right);
        strcpy(lastRotation, "Right-Left rotation (RL)");
        return leftRotate(node);
    }
    if(strcmp(lastRotation,"")==0) strcpy(lastRotation,"No rotation");
    return node;
}

void printBF(Node* root){
    if(!root) return;
    printBF(root->left);
    printf("Node %d: BF = %d\n", root->data, getBF(root));
    printBF(root->right);
}

/* interactive insertion: read count, then values; after each insertion print rotation and all BFs */
int main(){
    int n, x;
    Node* root = NULL;
    printf("How many insertions? ");
    if(scanf("%d",&n)!=1) return 0;
    printf("Enter values to insert (one per line):\n");
    for(int i=0;i<n;i++){
        scanf("%d",&x);
        lastRotation[0]=0;
        root = insert(root, x);
        if(strlen(lastRotation)==0) strcpy(lastRotation,"No rotation");
        printf("After inserting %d -> Rotation: %s\n", x, lastRotation);
        printf("Current nodes' balance factors:\n");
        printBF(root);
        printf("-------\n");
    }
    return 0;
}
```

-  Height of the tree

```python
%%writefile 10_avl_deletion.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct Node { int data; int height; struct Node *left,*right; } Node;
Node* newNode(int v){ Node* n=malloc(sizeof(Node)); n->data=v; n->height=1; n->left=n->right=NULL; return n; }
int height(Node* n){ return n ? n->height : 0; }
int getBF(Node* n){ return n ? height(n->left) - height(n->right) : 0; }
char lastRotation[128];

Node* rightRotate(Node* y){
    Node* x = y->left;
    Node* T2 = x->right;
    x->right = y;
    y->left = T2;
    y->height = 1 + (height(y->left) > height(y->right) ? height(y->left) : height(y->right));
    x->height = 1 + (height(x->left) > height(x->right) ? height(x->left) : height(x->right));
    strcpy(lastRotation, "Right rotation (LL)");
    return x;
}

Node* leftRotate(Node* x){
    Node* y = x->right;
    Node* T2 = y->left;
    y->left = x;
    x->right = T2;
    x->height = 1 + (height(x->left) > height(x->right) ? height(x->left) : height(x->right));
    y->height = 1 + (height(y->left) > height(y->right) ? height(y->left) : height(y->right));
    strcpy(lastRotation, "Left rotation (RR)");
    return y;
}

Node* minValueNode(Node* node){
    Node* cur = node;
    while(cur->left) cur = cur->left;
    return cur;
}

Node* insert(Node* node, int key){
    if(!node) { return newNode(key); }
    if(key < node->data) node->left = insert(node->left, key);
    else if(key > node->data) node->right = insert(node->right, key);
    else return node;
    node->height = 1 + (height(node->left) > height(node->right) ? height(node->left) : height(node->right));
    int bf = getBF(node);
    // LL
    if(bf > 1 && key < node->left->data) return rightRotate(node);
    // RR
    if(bf < -1 && key > node->right->data) return leftRotate(node);
    // LR
    if(bf > 1 && key > node->left->data){
        node->left = leftRotate(node->left);
        strcpy(lastRotation, "Left-Right rotation (LR)");
        return rightRotate(node);
    }
    // RL
    if(bf < -1 && key < node->right->data){
        node->right = rightRotate(node->right);
        strcpy(lastRotation, "Right-Left rotation (RL)");
        return leftRotate(node);
    }
    return node;
}

Node* deleteNode(Node* root, int key){
    if(!root) { strcpy(lastRotation,"No rotation (not found)"); return root; }
    if(key < root->data) root->left = deleteNode(root->left, key);
    else if(key > root->data) root->right = deleteNode(root->right, key);
    else {
        if(!root->left || !root->right){
            Node* temp = root->left ? root->left : root->right;
            if(!temp){ temp = root; root = NULL; free(temp); }
            else { *root = *temp; free(temp); }
        } else {
            Node* temp = minValueNode(root->right);
            root->data = temp->data;
            root->right = deleteNode(root->right, temp->data);
        }
    }
    if(!root) return root;
    root->height = 1 + (height(root->left) > height(root->right) ? height(root->left) : height(root->right));
    int bf = getBF(root);

    // Balancing
    if(bf > 1 && getBF(root->left) >= 0) return rightRotate(root); // LL
    if(bf > 1 && getBF(root->left) < 0){
        root->left = leftRotate(root->left);
        strcpy(lastRotation, "Left-Right rotation (LR)");
        return rightRotate(root);
    }
    if(bf < -1 && getBF(root->right) <= 0) return leftRotate(root); // RR
    if(bf < -1 && getBF(root->right) > 0){
        root->right = rightRotate(root->right);
        strcpy(lastRotation, "Right-Left rotation (RL)");
        return leftRotate(root);
    }
    strcpy(lastRotation, "No rotation");
    return root;
}

void preorder(Node* root){ if(!root) return; printf("%d ", root->data); preorder(root->left); preorder(root->right); }
int treeHeight(Node* root){ return root ? root->height : 0; }

int main(){
    Node* root = NULL;
    int n, x, m;
    printf("First insert initial nodes into AVL tree. How many? ");
    if(scanf("%d",&n)!=1) return 0;
    printf("Enter %d values:\n", n);
    for(int i=0;i<n;i++){ scanf("%d",&x); root = insert(root,x); }
    printf("Initial preorder: "); preorder(root); printf("\nInitial height: %d\n", treeHeight(root));
    printf("How many deletions to perform? ");
    scanf("%d",&m);
    printf("Enter deletion keys one per line:\n");
    for(int i=0;i<m;i++){
        scanf("%d",&x);
        lastRotation[0]=0;
        root = deleteNode(root, x);
        if(strlen(lastRotation)==0) strcpy(lastRotation,"No rotation");
        printf("After deleting %d -> Rotation: %s\n", x, lastRotation);
        printf("Current tree height: %d\n", treeHeight(root));
        printf("Preorder now: "); preorder(root); printf("\n-------\n");
    }
    return 0;
}
```
